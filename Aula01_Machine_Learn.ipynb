```python
# [Índice 1] Pergunta: O animal voa?
# [Índice 2] Pergunta: O animal respira de baixo d'água?
# [Índice 3] Pergunta: O animal está vivo?

#Cada linha representa um animal.
#Cada coluna representa uma característica (feature).
#O vetor [0,1,1] é uma linha com três valores binários que representam as respostas às perguntas.

peixe1 = [0,1,1]
peixe2 = [0,1,1]
peixe3 = [0,1,0]
ave1 = [1,0,0]
ave2 = [1,0,1]
ave3 = [1,0,1]

# Rótulo: 1 = Peixe, 0 = Ave
# teste_x: são os vetores de características (features) dos animais.
teste_x = [peixe1, peixe2, peixe3, ave1, ave2, ave3]

#teste_y: são os rótulos (labels), indicando se o animal é peixe (1) ou ave (0).
teste_y = [1,1,1,0,0,0]
```

```python
# Exportando o modelo que irei utilizar para este problema.
from sklearn.svm import LinearSVC
modelo = LinearSVC()
# Eu agora vou definir o que será acrescentado na base de conhecimento.
# O teste_y está servindo como uma maneira de criar o cenário de aprendizado segmentado para a máquina.
modelo.fit(teste_x, teste_y)
```

```python
# Agora será criada a etapa de testes para previsões de máquina.
misterio1 = [1,0,1]
misterio2 = [0,1,1]
misterio3 = [1,0,1]
misterio4 = [1,0,1]

# Portanto nessa análise, descobriremos a partir da classificação binária, em qual das classes se encontram os novos elementos criados que estão fora do conhecimento da máquina.
testes = [misterio1, misterio2, misterio3, misterio4]
# As previsões podem ser lidas como a quantidade de questões que vão existir em uma prova, para cada um delas podemos acertar ou errar.
previsoes = modelo.predict(testes)
```

```python
# Lista com os rótulos corretos dos dados de teste. (Eu já irei acrescentar qual é a resposta esperada)
# Serve para comparar com as previsões feitas pelo modelo e avaliar sua precisão. (Será como o processo de corrigir uma prova)
testes_classes_corretas = [0,1,0,0]
```

```python
resultado_avaliacao = (previsoes == testes_classes_corretas).sum()
total = len(testes)
taxa_de_acerto = resultado_avaliacao/total * 100
print(f"Taxa de acerto:{taxa_de_acerto:}%")
```

```python
from sklearn.metrics import accuracy_score
taxa_de_acerto = accuracy_score(testes_classes_corretas, previsoes)
print(f"Taxa de acerto:{taxa_de_acerto * 100}%")
```
